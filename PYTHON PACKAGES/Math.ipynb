```python
import math

number = 25
sqrt_result = math.sqrt(number)
print(f"Square root of {number}: {sqrt_result}")

decimal_up = 4.2
ceil_result = math.ceil(decimal_up)
print(f"Ceiling of {decimal_up}: {ceil_result}")

decimal_down = 4.8
floor_result = math.floor(decimal_down)
print(f"Floor of {decimal_down}: {floor_result}")

base = 2
exponent = 3
pow_result = math.pow(base, exponent)
print(f"{base} raised to the power of {exponent}: {pow_result}")

fact_number = 5
factorial_result = math.factorial(fact_number)
print(f"Factorial of {fact_number}: {factorial_result}")
```
