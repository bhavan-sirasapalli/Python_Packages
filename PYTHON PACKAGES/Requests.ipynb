```python
import requests

# 1. requests.get(): Make a GET request
# This fetches data from a specified URL.
try:
    response_get = requests.get("https://api.github.com/events")
    print(f"GET Request Status Code: {response_get.status_code}")
    # 2. Access JSON content
    if response_get.status_code == 200:
        print(f"First event type (from JSON): {response_get.json()[0]["type"]}")
except requests.exceptions.RequestException as e:
    print(f"GET Request Error: {e}")

# 3. requests.post(): Make a POST request
# This sends data to a specified URL.
# Using a dummy API for demonstration.
post_data = {"title": "foo", "body": "bar", "userId": 1}
try:
    response_post = requests.post("https://jsonplaceholder.typicode.com/posts", json=post_data)
    print(f"POST Request Status Code: {response_post.status_code}")
    if response_post.status_code == 201:
        print(f"POST Response (from JSON): {response_post.json()}")
except requests.exceptions.RequestException as e:
    print(f"POST Request Error: {e}")
```
