```python
import matplotlib.pyplot as plt
import numpy as np

x_line = np.linspace(0, 10, 100)
y_line = np.sin(x_line)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(x_line, y_line, label='sin(x)', color='blue')
plt.title('Line Plot: sin(x)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()


x_scatter = np.random.rand(50) * 10
y_scatter = 2 * x_scatter + np.random.randn(50) * 5

plt.subplot(1, 3, 2)
plt.scatter(x_scatter, y_scatter, color='green', alpha=0.7, s=50)
plt.title('Scatter Plot: Random Data')
plt.xlabel('X-value')
plt.ylabel('Y-value')
plt.grid(True)


data_hist = np.random.normal(loc=0, scale=1, size=1000)

plt.subplot(1, 3, 3)
plt.hist(data_hist, bins=30, color='red', alpha=0.7, edgecolor='black')
plt.title('Histogram: Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)

plt.tight_layout()
plt.show()
```
