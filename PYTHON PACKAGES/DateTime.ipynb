```python
import datetime

now = datetime.datetime.now()
print(f"Current date and time: {now}")

today = datetime.date.today()
print(f"Current date: {today}")

specific_date_time = datetime.datetime(2023, 10, 27, 10, 30, 0)
print(f"Specific date and time: {specific_date_time}")

specific_date = datetime.date(2023, 1, 15)
print(f"Specific date: {specific_date}")

specific_time = datetime.time(14, 45, 20)
print(f"Specific time: {specific_time}")


formatted_dt = now.strftime("%Y-%m-%d %H:%M:%S")
print(f"Formatted date and time: {formatted_dt}")

formatted_date = today.strftime("%A, %B %d, %Y")
print(f"Formatted date: {formatted_date}")

one_week_later = now + datetime.timedelta(days=7)
print(f"One week later: {one_week_later}")

two_hours_ago = specific_date_time - datetime.timedelta(hours=2)
print(f"Two hours before specific time: {two_hours_ago}")

time_difference = one_week_later - now
print(f"Time difference (timedelta object): {time_difference}")
print(f"Time difference in days: {time_difference.days}")

year = now.year
month = now.month
day = now.day
hour = now.hour
minute = now.minute
second = now.second
print(f"Year: {year}, Month: {month}, Day: {day}, Hour: {hour}, Minute: {minute}, Second: {second}")

date_string = "2024-03-15 11:00:00"
parsed_dt = datetime.datetime.strptime(date_string, "%Y-%m-%d %H:%M:%S")
print(f"Parsed datetime from string: {parsed_dt}")

date_only_string = "January 20, 2023"
parsed_date = datetime.datetime.strptime(date_only_string, "%B %d, %Y").date()
print(f"Parsed date from string: {parsed_date}")
```
