```python
import tkinter as tk
from tkinter import messagebox

# 1. Create the main application window
app = tk.Tk()
app.title("Tkinter Features Demo")
app.geometry("400x250") # Set window size

# Function to be called by the button
def on_button_click():
    user_input = entry_field.get() # Get text from entry field
    if user_input:
        messagebox.showinfo("Greeting", f"Hello, {user_input}!")
    else:
        messagebox.showwarning("Input Missing", "Please enter your name!")

# 2. Add a Label widget
label = tk.Label(app, text="Enter your name:", font=("Arial", 12))
label.pack(pady=10)

# 3. Add an Entry widget
entry_field = tk.Entry(app, width=30, font=("Arial", 12))
entry_field.pack(pady=5)

# 4. Add a Button widget
button = tk.Button(app, text="Say Hello", command=on_button_click, font=("Arial", 12))
button.pack(pady=10)

# Run the application
app.mainloop()
```
