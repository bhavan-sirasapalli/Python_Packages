```python
import os

# 1. os.getcwd(): Get the current working directory
current_directory = os.getcwd()
print(f"Current working directory: {current_directory}")

# 2. os.listdir(): List contents of a directory
# This will list files and directories in the current working directory
print(f"Contents of current directory: {os.listdir(current_directory)}")

# 3. os.path.join(): Join path components intelligently
file_name = "my_document.txt"
full_path = os.path.join(current_directory, file_name)
print(f"Joined path: {full_path}")
```
